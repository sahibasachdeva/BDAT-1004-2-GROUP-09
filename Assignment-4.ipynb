The algorithm then takes the new first number in L (3) and appends it to list primeL, and then removes from L 3 and all its remaining multiples (9,15,21,...). So, in every iteration, the first number of list L is appended to list primeL and then it and its multiples are removed from list L. The iterations stop when list L becomes empty. Write a function sieve() that takes as input a positive integer n, implements the above algorithm, and returns a list of all prime numbers up to n.

```python
import math

def sieve(number):
    
    primes = []   
    for i in range(2,number+1):
        primes.append(i)
    i = 2
    while(i <= int(math.sqrt(number))):
        if i in primes:
            for j in range(i*2, number+1, i):
                if j in primes:
                    primes.remove(j)
        i = i+1
    print (primes)


print ("Enter the number until you want prime numbers ")
number1 = int(input())
sieve(number1)
```

#### 7.Two words are anagrams if the letters of one can be rearranged to form the other. For example, mar, ram, and arm are all anagrams of each other. One way to quickly identify whether two arbitrary words are anagrams are to sort the letters in each word to produce a signature. (The above words all have the signature 'amr'.) Two words are anagrams if and only if they have the same signature. Write a function called anagrams() that takes the name of a text file as its only parameter. The file is assumed to have one word per line, all lower-case. The function returns a dictionary in which the keys are signatures of words in the file and the value for each key is the set of all words in the file having that particular signature.

```python
def sig(s):
    w = list(s)
    w.sort()
    w = ''.join(w)
    return w

def anagrams(x):
    y = {}
    for line in open(x):
        word = line.strip().lower()
        w = sig(word)

        if w not in y:
            y[w] = [word]
        else:
            y[w].append(word)

    for i in y :
        print(i, y[i])
```

#### 10. Write a function wordcount() that takes the name of a text file as input and prints the number of occurrences of every word in the file. You function should be case-insensitive so 'Hello' and 'hello' are treated as the same word. You should ignore words of length 2 or less. Test your implementation on file frankenstein.txt.

```python
def wordcount(inputFile):
    wordsCount = dict()
    for words in inputFile:
        words = words.strip()
        words = words.split(" ")    
        for word in words:
            if len(word) > 2:
                if word in wordsCount:
                    wordsCount[word] = wordsCount[word] + 1
                else:
                    wordsCount[word] = 1

    for x,y in wordsCount.items():
        print(x,y)
    
file = input("enter the file for wordcount")
inputFile = open(file, "r")
wordcount(inputFile)
```
